#### The challenge proposes the following situation:
####Imagine that you have been hired by a large bank to develop its new system. This bank wants to modernise its operations and has chosen the Python language. For the first version of the system we must implement only 3 operations: deposit, withdrawal and statement.

####<u>Rules to be respected:</u>

#### **Deposit operations**
####It must be possible to deposit positive amounts to my bank account. Version 1 of the project only works with 1 user, so there is no need to worry about identifying which is the branch and bank account number. All deposits must be stored in a variable and displayed in the statement operation.

#### **Withdrawal operations**
####The system must allow 3 daily withdrawals with a maximum limit of R$ 500.00 per withdrawal. If the user does not have a balance in the account, the system must display a message stating that it will not be possible to withdraw the money due to lack of balance. All withdrawals must be stored in a variable and displayed in the statement operation.

#### **Transaction operations**
####This operation should list all deposits and withdrawals made from the account. At the end of the listing the current balance of the account should be displayed. If the statement is blank, display the message: No transactions were made.
####The amounts must be displayed using the format R$ xxx.xx.

```python
menu = """
[d] deposit
[w] withdraw
[s] statement
[e] exit

=> """
balance = 0
limit = 500
statement = ""
number_withdraw = 0
WITHDRAW_LIMIT = 2

while True:
    option = input(menu)

    if option == "d":
        amount = float(input("Inform the deposit amount: "))

        if amount > 0:
            balance += amount
            statement += f'Deposit: R$ {amount:.2f}\n'
        else:
            print("Operation failed! The value entered is invalid. Please try again!")

    elif option == 'w':
        amount = float(input("Inform the withdrawal amount: "))

        exceeded_balance = amount > balance
        exceeded_limit = amount > limit
        exceeded_withdraw = number_withdraw >= WITHDRAW_LIMIT

        if exceeded_balance:
            print("Operation failed! You do not have enough money.")
        elif exceeded_limit:
            print("Operation failed! The withdrawal amount exceeded the limit.")
        elif exceeded_withdraw:
            print("Operation failed! Number of withdrawals was exceeded.")
        elif amount > 0:
            balance -= amount
            statement += f'Withdraw: R$ {amount:.2f}\n'
            number_withdraw += 1
        else:
            print("Operation failed! The value entered is invalid.")

    elif option == "s":
        print("\n*************Bank Statement*************")
        print("No transactions were carried out." if not statement else statement)
        print(f"\nbalance: R$ {balance:.2f}")
        print("****************************************")

    elif option == "e":
        break

    else:
        print("Invalid operation, please reselect the desired operation.")
```
